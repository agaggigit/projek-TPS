Used Parameter:
<br>
$\lambda =$ Arrival Rate $= \
{1}{\text{rata-rata interarrival time}}$
<br>
$\mu =$ Service Rate
<br>
$\rho =$ Utilization $= \frac{\lambda}{c \times \mu}$
<br>
$Lq =$ Length of Queue
<br>
$Wq =$ Waiting time in Queue

```python
df_rekap_nodash = df_rekap.replace('-', np.nan)

avg_itrarrival = df_rekap_nodash['Interarrival Time (s)'].astype(float)
nozzle_count = df_rekap_nodash['Jumlah Nozzle Aktif'].astype(float)

session = df_rekap_nodash['Sesi']
arrival_rate = 1/avg_itrarrival
service_rate = df_rekap_nodash['Rerata Waktu Pelayanan (s)'].astype(float)
utilization = arrival_rate/(nozzle_count*service_rate)
Lq = df_rekap_nodash['Rerata Panjang Antrean (s)'].astype(float)
Wq = df_rekap_nodash['Rerata Waktu Mengantri (s)'].astype(float)

df_parameters = pd.DataFrame([session, arrival_rate, service_rate, utilization, Lq, Wq])
df_parameters = pd.DataFrame({
    'Sesi': df_rekap_nodash['Sesi'],
    'λ (Lambda)': arrival_rate,
    'μ (Mu)': service_rate,          
    'ρ (Rho)': utilization,                 
    'Lq': df_rekap_nodash['Rerata Panjang Antrean (s)'],
    'Wq': df_rekap_nodash['Rerata Waktu Mengantri (s)']
})
df_parameters
```
